```python
# Import necessary libraries
import dash
from dash import dcc, html, Input, Output
import spacy

# Load spaCy language model for similarity calculation
nlp = spacy.load("en_core_web_sm")

# Initialize the Dash app
app = dash.Dash(__name__)
```

```python
# Define the layout of the app with sky blue background
app.layout = html.Div([
    # Title at the top
    html.H1("Text Similarity Calculator", style={'textAlign': 'center', 'color': 'darkblue'}),

    # First text input field
    dcc.Input(
        id="text1",
        type="text",
        placeholder="Enter the first text",
        style={'width': '100%', 'marginTop': '10px'}
    ),

    # Second text input field
    dcc.Input(
        id="text2",
        type="text",
        placeholder="Enter the second text",
        style={'width': '100%', 'marginTop': '10px'}
    ),

    # Submit button
    html.Button(
        'Calculate Similarity',
        id='submit-button',
        n_clicks=0,
        style={'marginTop': '10px', 'backgroundColor': 'lightblue'}
    ),

    # Output area for displaying similarity score
    html.Div(id="output", style={'marginTop': '20px', 'fontSize': '20px'})
],
style={'backgroundColor': 'skyblue', 'padding': '20px', 'borderRadius': '10px'}
)

# Define callback to calculate and display similarity score
@app.callback(
    Output("output", "children"),
    Input("submit-button", "n_clicks"),
    Input("text1", "value"),
    Input("text2", "value")
)
def calculate_similarity(n_clicks, text1, text2):
    # Only calculate if button is clicked and both texts are provided
    if n_clicks > 0 and text1 and text2:
        doc1 = nlp(text1)
        doc2 = nlp(text2)
        similarity_score = doc1.similarity(doc2)
        return f"Similarity Score: {similarity_score:.2f}"
    return ""

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)
```
